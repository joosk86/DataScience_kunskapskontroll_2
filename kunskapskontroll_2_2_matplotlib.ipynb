### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure = The whole figure. The Figure keeps track of all the axes.
# Axes = An Axes is an Artist attached to a Figure that contains a region for plotting data, and usually includes two (or three in the case of 3D).
# Axis = These objects (x-axis, y-axis etc) set the scale and limits and generate ticks (the marks on the Axis).
# Artist = Basically, everything visible on the Figure is an Artist (even Figure, Axes, and Axis objects).
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# OO approach
x = np.linspace(-4, 4, 100)

y = x**2

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, y, label='$y = x^2$', color = "black")
ax.set_xlabel('x')
ax.set_ylabel('y', rotation = 0)
ax.set_title("Johns Plot")
ax.legend()
```

```python
# Pyplot approach
x = np.linspace(-4, 4, 100)

y = x**2

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, y, label='$y = x^2$')
plt.xlabel('x')
plt.ylabel('y', rotation = 0)
plt.title("Johns Plot")
plt.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python

fig, axs = plt.subplot_mosaic([['left', 'right']], layout='constrained', figsize=(12, 5))

                              
axs['left'].set_title('Scatter plot', fontsize=14,)
axs['right'].set_title('Bar plot', fontsize=14,)

axs["left"].scatter(random_data_x, random_data_y, color="purple",)
axs["left"].set_xlabel('x', fontsize=12)
axs["left"].set_ylabel('y', fontsize=12, rotation = 0)
axs["left"].plot(x, y, label='$y = x^2$', color = "black")
axs["left"].legend(loc=3)

axs["right"].bar(names, values, color="green", label="$Sales$")
axs["right"].set_xlabel('Fruit', fontsize=10)
axs["right"].set_ylabel('Quantity', fontsize=10, rotation = 0)
axs["right"].legend()


plt.suptitle("Scatter Plot and Bar Plot", fontsize=16, weight="bold")
```
